```python
class Node:
    def __init__(self, data):
        self.data = data
        self.nref = None
        self.pref = None

class doublyLL:
    def __init__(self):
        self.head = None

    def print_dll(self):
        if self.head is None:
            print("LL is empty")
        else:
            n = self.head
            while n is not None:
                print(n.data, "-->", end=" ")
                n = n.nref

    def reverse_print(self):
        print()
        if self.head is None:
            print("Linked is empty")
        else:
            n = self.head
            while n.nref is not None:
                n = n.nref
            while n is not None:
                print(n.data, "-->", end=" ")
                n = n.pref

    def insert_empty(self, data):
        if self.head is None:
            new_node = Node(data)
            self.head = new_node
        else:
            print("Linked List is not empty")

    def add_begin(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            new_node.nref = self.head
            self.head.pref = new_node
            self.head = new_node

    def add_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            n = self.head
            while n.nref is not None:
                n = n.nref
            n.nref = new_node
            new_node.pref = n

    def add_after_value(self, existing_value, data):
        new_node = Node(data)
        if self.head is None:
            print("Linked List is empty")
        else:
            n = self.head
            while n is not None:
                if n.data == existing_value:
                    new_node.nref = n.nref
                    if n.nref is not None:
                        n.nref.pref = new_node
                    n.nref = new_node
                    new_node.pref = n
                    if new_node.nref is not None:
                        new_node.nref.pref = new_node  # Update the pref reference of the node after the new node
                    break
                n = n.nref

    def delete_begin(self):
        if self.head is None:
            print("Linked List is empty, cannot delete from the beginning")
        else:
            self.head = self.head.nref
            if self.head is not None:
                self.head.pref = None

    def delete_by_value(self, value):
        if self.head is None:
            print("Linked List is empty, cannot delete by value")
        else:
            n = self.head
            while n is not None:
                if n.data == value:
                    if n.pref is not None:
                        n.pref.nref = n.nref
                    if n.nref is not None:
                        n.nref.pref = n.pref
                    break
                n = n.nref

    def delete_end(self):
        if self.head is None:
            print("Linked List is empty, cannot delete from the end")
        else:
            n = self.head
            while n.nref is not None:
                n = n.nref
            if n.pref is not None:
                n.pref.nref = None
            else:
                self.head = None

# Example usage:

dll = doublyLL()

dll.add_begin(3)
dll.add_begin(2)
dll.add_begin(1)

print("Doubly Linked List:")
dll.print_dll()

print("\nReverse Print:")
dll.reverse_print()

dll.add_end(4)
print("\nDoubly Linked List after adding at the end:")
dll.print_dll()

dll.add_after_value(2, 5)
print("\nDoubly Linked List after adding after a specific value:")
dll.print_dll()

dll.delete_begin()
print("\nDoubly Linked List after deleting from the beginning:")
dll.print_dll()

dll.delete_by_value(4)
print("\nDoubly Linked List after deleting by value:")
dll.print_dll()

dll.delete_end()
print("\nDoubly Linked List after deleting from the end:")
dll.print_dll()
```

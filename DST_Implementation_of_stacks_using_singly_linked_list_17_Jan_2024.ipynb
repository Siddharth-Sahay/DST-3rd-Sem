```python
class Node:
  def __init__(self,x):
    self.data  = x
    self.next = None

class stack:
  def __init__(self):
    self.top = None
  def push(self,limit):
    if self.top == None:
      num = int(input("Enter the number you want to push - "))
      new = Node(num)
      self.top= new
      self.top.next = None
    else:
      count = 0
      temp = self.top
      while temp is not None:
        count = count + 1
        temp = temp.next
      if limit > count:
        num = int(input("Enter the number you want to push - "))
        new = Node(num)
        new.next = self.top
        self.top = new
      else:
        print("Stack Overflow")
        print("--------------------------------------")
  def popn(self):
    if self.top is None:
      print("The stack is empty")
      print("--------------------------------------")
    elif self.top.next is None:
      print("poped element is - ",self.top.data)
      print("--------------------------------------")
      self.top = None
    else:
      print("Poped element is - ",self.top.data)
      print("--------------------------------------")
      self.top = self.top.next

  def print_stack(self):
    if self.top is None:
      print("Stack is empty")
      print("--------------------------------------")
    else:
      n = self.top
      while n is not None:
        print ("--> ",n.data,"\n")
        n = n.next
      print("--------------------------------------")


s = stack()
limit = int(input("What is the limit of the stack - "))
while True:
  print("Choose the options below")
  print("1.Push \n2.Pop \n3.Display \n4.Exit")
  optn = int(input())
  print("--------------------------------------")
  if optn == 1:
    print("PUSH OPERATION")
    s.push(limit)
    print("--------------------------------------")
  elif optn == 2:
    print("POP OPERATION")
    s.popn()
    print("--------------------------------------")
  elif optn == 3:
    print("DISPLAY OPERATION")
    s.print_stack()
    print("--------------------------------------")
  elif optn == 4:
    print(" EXITINHG")
    break
  else:
    print("ENTER A THE CORREXT OPTION")




```
